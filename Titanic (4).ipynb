```python
data.drop("Cabin", axis=1, inplace=True)
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
sns.heatmap(data.isnull(), yticklabels=False, cbar=False)
```

```python
data.isnull().sum()
```

```python
sex=pd.get_dummies(data['Sex'],drop_first=True)
gender=sex.astype(int)
gender.head(5)
```

```python
emb=pd.get_dummies(data['Embarked'], drop_first=True)
embrak=emb.astype(int)
embrak.head(5)
```

```python
pcl=pd.get_dummies(data['Pclass'], drop_first=True)
Pcl=pcl.astype(int)
Pcl.head(5)
```

```python
data1=pd.concat([data, gender, embrak, Pcl], axis=1)
data1.head(5)
```

```python
data1.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Embarked'], axis=1, inplace=True)
data1.head(5)
```

```python
data1.drop(['Pclass'], axis=1, inplace=True)
data1.head(5)
```

# <span style="color:pink;" >Train Data & Test Data</span>

```python
X=data1.drop('Survived',axis=1)
y=data1["Survived"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
X_train.columns = X_train.columns.astype(str)

model.fit(X_train, y_train)
```

```python
X_test.columns=X_test.columns.astype(str)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)

# Visualize the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix',c='red')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

From above, accuracy is almost 75% which is quite good,we cann't say that accuracy is bad
